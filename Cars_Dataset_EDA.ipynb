```python
#Continuing with other columns now.
#Torque is meaured in NM (Newton meters). rpms are not necessary for torque. As the data constains majority values only in Nm, i am
#replacing @digitsrpm with ' ' in order to maintain consistent entries in the column. regex=True is used to set the pattern for future versions as well. Replacing all entries of @(/d) for consistency.
df['Torque(Nm)']=df['Torque(Nm)'].astype(str).apply(lambda t:t.replace('Nm', ''))
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@\d+rpm', ' ', regex=True).str.strip()
df['Torque(Nm)']=df['Torque(Nm)'].astype(str).apply(lambda o:o.replace('nm', ''))
df['Torque(Nm)']=df['Torque(Nm)'].astype(str).apply(lambda u:u.replace('NM', ''))
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@\d+RPM', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@(\d+-\d+)rpm', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@(\d+-\d+)RPM', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@\d+ rpm', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@(\d+-\d+)', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@ \d+ RPM', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@ \d+ rpm', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@(\d+ - \d+)rpm', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@ (\d+-\d+) rpm', ' ', regex=True).str.strip()
df['Torque(Nm)'] = df['Torque(Nm)'].str.replace(r'@\d+', ' ', regex=True).str.strip()
```

```python
#Finding co-orelation between desired numerical columns!Correlation ranges from -1 to +1. Here i am just providing an overlay on the correlation matrix in form of a heatmap.
#The correlation matrix is a square matrix that measures the strength & direction of a linear relationship.
numerical_columns = ['Cylinders', 'Displacement(cc)', 'Torque(Nm)']
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="rocket")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
#Let's take one final plot and complete our EDA - A cluster map using seaborn. It is basically a clustered heatmap for numerical columns. A heatmap, as we know, represents the correlation matrix b/w columns.
#Cluster map refers to hierarchal clustering applied to rows and columns of a hatmap based on their similarity in terms of correlation. the colors here represent the strength of correlation.
#A darker color indicates stronger correlation( can be either +ve or -ve )!
numerical_columns = ['Price', 'Displacement(cc)', 'Torque(Nm)']
plt.figure(figsize=(10,6))
sns.clustermap(df[numerical_columns].corr(), annot=True, cmap='flare')
plt.show()
```
